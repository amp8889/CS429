```python
# Information taken from:
# ChatGPT
# https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_20newsgroups.html
# https://www.tutorialspoint.com/python_text_processing/python_tokenization.htm
# https://www.geeksforgeeks.org/nlp-how-tokenizing-text-sentence-words-works/

from sklearn.datasets import fetch_20newsgroups
import string



newsGroupsData = fetch_20newsgroups(data_home=None, subset='all', categories=None, shuffle=False, remove=(), download_if_missing=True, return_X_y=False)


def tokenizeDocument(document):

    translator = str.maketrans('', '', string.punctuation)
    document = document.translate(translator)
    terms = document.split()
    return terms


documentsToken = [tokenizeDocument(doc) for doc in newsGroupsData.data]

# print("Original:")
# print(newsGroupsData.data[2])

# print("\nTokenized Words:")
# print(documentsToken[2])

# print ("------------------------------------------------------------------------")


# print("Original:")
# print(newsGroupsData.data[5553])

# print("\nTokenized Words:")
# print(documentsToken[5552])
# print ("------------------------------------------------------------------------")


# print("Original:")
# print(newsGroupsData.data[25])

# print("\nTokenized Words:")
# print(documentsToken[25])




```

```python
# Information and code borrowed from these sources:
# https://www.geeksforgeeks.org/inverted-index/
# ChatGPT
# https://codereview.stackexchange.com/questions/281874/simple-inverted-index
# https://github.com/Noureldin2303/Inverted-index-python



def createInvertedIndex(documentsToken):
    invertedIndex = {}

    for doc_id, terms in enumerate(documentsToken):
        for term in set(terms):  

            if term in invertedIndex:
                invertedIndex[term].append(doc_id)

            else:
                invertedIndex[term] = [doc_id]


    for term in invertedIndex:

        invertedIndex[term] = sorted(invertedIndex[term])

    return invertedIndex




invertedIndex = createInvertedIndex(documentsToken)


print("Inverted Index for 'chicago':", invertedIndex.get('chicago', []))
print("Inverted Index for 'bears':", invertedIndex.get('bears', []))

print("Inverted Index for 'school':", invertedIndex.get('school', []))
print("Inverted Index for 'teachers':", invertedIndex.get('teachers', []))

print("Inverted Index for 'hospital':", invertedIndex.get('hospital', []))
print("Inverted Index for 'drugs':", invertedIndex.get('drugs', []))
```

```python
# Info and Code borrowed From
# https://www.geeksforgeeks.org/python-intersection-two-lists/
# ChatGPT
# https://www.w3schools.com/python/ref_set_intersection.asp


def intersectLists(listOne, listTwo):

    result = []
    i = 0
    j = 0

    while i < len(listOne) and j < len(listTwo):
        
        if listOne[i] == listTwo[j]:

            result.append(listOne[i])
            i += 1
            j += 1

        elif listOne[i] < listTwo[j]:
            i += 1

        else:
            j += 1


    return result



def wordSort(words, index):

    sortedWords = sorted(words, key=lambda word: len(index.get(word, [])))
    return sortedWords



resultIntersection = intersectLists(invertedIndex.get('school', []), invertedIndex.get('teacher', []))
print("Intersection of 'school' and 'teacher':", resultIntersection)

resultIntersection = intersectLists(invertedIndex.get('hospital', []), invertedIndex.get('drugs', []))
print("Intersection of 'hospital' and 'drugs':", resultIntersection)



wordsToBeSorted = ['hospital', 'drugs', 'school', 'teacher', 'chicago', 'bears']
sortedWords = wordSort(wordsToBeSorted, invertedIndex)
print("Sorted words:", sortedWords)
```

```python
#Information taken from
# Chatgpt 
# https://www.geeksforgeeks.org/inverted-index/
# https://medium.com/swlh/demystifying-a-web-search-problem-using-inverted-index-c6df8236291


def search(query, invertedIndex):

    queryTerms = tokenizeDocument(query)
    sortedQueryTerms = wordSort(queryTerms, invertedIndex)
    result = invertedIndex.get(sortedQueryTerms[0], [])

    for term in sortedQueryTerms[1:]:
        
        result = intersectLists(result, invertedIndex.get(term, []))

    return result





query = "school AND teacher"
searchResults = search(query, invertedIndex)
print("Results for this query '{}':".format(query))
print(searchResults)


query = "hospital AND drugs"
searchResults = search(query, invertedIndex)
print("Results for this query '{}':".format(query))
print(searchResults)


query = "Chicago AND news"
searchResults = search(query, invertedIndex)
print("Results for this query '{}':".format(query))
print(searchResults)

```
